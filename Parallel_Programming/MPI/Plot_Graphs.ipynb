```python
import numpy as np
import matplotlib.pyplot as plt

file1 = open(r"table_out.txt", "r")

lines = file1.read().splitlines()
s1 = np.array([float(line.strip()[8:]) for line in lines[:8]])
s2 = np.array([float(line.strip()[11:]) for line in lines[8:16]])
s3 = np.array([float(line.strip()[13:]) for line in lines[16:24]])
plt.figure(figsize=(14,8))
plt.plot(np.linspace(1,8,8), s1, 'orange', label = "N = 1000")
plt.plot(np.linspace(1,8,8), s2, 'b', label = "N = 10^6")
plt.plot(np.linspace(1,8,8), s3, 'red',label = "N = 10^8")
plt.xlabel(r'Processes', fontsize = 20)
plt.ylabel(r'S', fontsize = 20)
plt.legend(fontsize = 14)
plt.show()

file1.close()
```
